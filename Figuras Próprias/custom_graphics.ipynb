```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_polynomial(polynomial_coeffs, points=None, gray_polynomial_coeffs=None):
    x = np.linspace(0, 3.3, 100)  
    
    y = np.polyval(polynomial_coeffs, x)
    
    plt.figure(figsize=(8, 6))

    plt.plot(x, y, label='Alvo real', color='blue')  # Blue color for the first polynomial
    
    if gray_polynomial_coeffs:
        y_gray = np.polyval(gray_polynomial_coeffs, x)
        plt.plot(x, y_gray, label='Previs√£o', color='gray', linestyle='--')  # Gray, dashed for the second polynomial
    
    if points:
        x_points, y_points = zip(*points)
        plt.scatter(x_points, y_points, color='green', label='Dados conhecidos', marker='o', s=100)
    
    plt.xlabel('Tempo')
    plt.ylabel('Valor')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
points = [(0.5, 0), (2.5, 0), (3.16, 1.8)]
coefficients = [1, -3, 1.25]
coefficients_2 =  [1, -7.2, 17.65, -17.25, 6.55, -0.75]
plot_polynomial(coefficients, points=points, gray_polynomial_coeffs=coefficients_2)
```

```python
def plot_error_surface(r):

    def error_function(x, y):
        return (np.sin(x) + np.sin(y) + np.cos(y) + np.cos(x/2) + (x**2)/10 + (y**2)/10 + x/5)

    x = np.linspace(-5, r, 100)
    y = np.linspace(-5, r, 100)
    x, y = np.meshgrid(x, y)
    z = error_function(x, y)

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Error')

    ax.view_init(30, 200)

    plt.show()
```

```python
plot_error_surface(10)
```
