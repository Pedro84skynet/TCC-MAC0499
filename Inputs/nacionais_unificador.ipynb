```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import csv
```

```python
df_imob = pd.read_csv('2014_01_imob.csv')
df_imob.set_index('Ano', inplace=True)
df_ipca = pd.read_csv('2014_01_ipca.csv')
df_ipca.set_index('Ano', inplace=True)
df_nfsp = pd.read_csv('2014_01_nfsp.csv')
df_nfsp.set_index('Ano', inplace=True)
df_selic = pd.read_csv('2014_01_selic.csv')
df_selic.set_index('Ano', inplace=True)
```

```python
df_nasc = pd.concat([df_imob, df_ipca, df_nfsp, df_selic], axis=1)
df_nasc
```

```python
df_nasc.to_csv('dados_nascionais.csv', index=True)
```
