```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import csv
import glob
```

```python
plt.rcParams['figure.max_open_warning'] = 91
```

```python
def plot_columns(df):
    colors = plt.cm.tab20(range(df.shape[1]))  

    for i, column in enumerate(df.columns):
        color = colors[i % 24]

        plt.figure()
        plt.plot(df.index, df[column], color=color)
        plt.xlabel('Ano')
        plt.ylabel('Valor')
        plt.title(column)

    plt.show()
```

```python
def concatenador():
    df_list = []
    csv_files = glob.glob("*.csv")
    for file in csv_files:
        df = pd.read_csv(file, index_col=0)  
        df_list.append(df)

    df_concatenated = pd.concat(df_list,axis=1)
    return df_concatenated
```

```python
df_input = concatenador()
df_input
```

```python
plot_columns(df_input)
```

```python
df_input.to_csv('2014_01_model_input.csv', index=True)
```
